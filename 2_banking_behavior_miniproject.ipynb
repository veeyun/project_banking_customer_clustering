```python
df_saving_tran_grouped = df_saving_tran.groupby('cust_id', as_index=False)

df_saving_tran_summary = pd.DataFrame()
df_saving_tran_summary["cust_id"]  = df_saving_tran_grouped.sum()["cust_id"]
df_saving_tran_summary["SV_tran_tot"]  = df_saving_tran_grouped.sum()["tran_amt"]
df_saving_tran_summary["SV_tran_mean"]  = df_saving_tran_grouped.mean()["tran_amt"]
df_saving_tran_summary["SV_tran_count"]  = df_saving_tran_grouped.count()["tran_amt"]
df_saving_tran_summary = df_saving_tran_summary[["cust_id", "SV_tran_tot", "SV_tran_mean", "SV_tran_count"]]
df_saving_tran_summary.head(10)

# helps us answer how much do people save and are they making lot of small transactions or few huge ones
```

```python
df_credit_tran_grouped = df_credit_tran.groupby('cust_id', as_index=False)

df_credit_tran_summary = pd.DataFrame()
df_credit_tran_summary["cust_id"]  = df_credit_tran_grouped.sum()["cust_id"]
df_credit_tran_summary["CC_tran_tot"]  = df_credit_tran_grouped.sum()["tran_amt"]
df_credit_tran_summary["CC_tran_avg"]  = df_credit_tran_grouped.mean()["tran_amt"]
df_credit_tran_summary["CC_tran_count"]  = df_credit_tran_grouped.count()["tran_amt"]
df_credit_tran_summary["CC_int_tot"]  = df_credit_tran_grouped.sum()["interest_amt"]
df_credit_tran_summary = df_credit_tran_summary[["cust_id", "CC_tran_tot", "CC_tran_avg", "CC_tran_count"]]
df_credit_tran_summary.head()

# helps us answer how much do people spend and are they making lot of small transactions or few huge ones
```
